## **Visão geral**

A AluraStore é uma rede formada por 4 lojas, de propriedade do Sr. João. Nestas lojas, os clientes podem encontrar à venda 51 itens diferentes que podem ser classificados em 8 categorias de produtos:

*   Móveis
*   Eletrodomésticos
*   Eletrônicos
*   Brinquedos
*   Utilidades domésticas
*   Livros
*   Instrumentos musicais
*   Esporte e lazer


Com vendas em todo o território brasileiro, o maior número de vendas é concentrado nos estados de São Paulo, Rio de Janeiro, Minas Gerais, Paraná e Rio Grande do Sul, dentre os quais São Paulo tem liderança isolada.

A rede aceita 4 formas de pagamento:

*   Cartão de crédito
*   Boleto
*   Cupom
*   Cartão de débito

sendo esta a frequência de escolha dos clientes.

Em sua maioria, as compras são parceladas, de 1 a 24, com concentração de 75% em até 4 parcelas.

## **Objetivo**

O Sr. João, proprietário da rede, tomou a decisão de vender uma das lojas, assim, capitalizando-a, para poder realizar um investimento em outro negócio.

Por isso, vamos analisar o desempenho das 4 lojas da rede entre os anos de 2020 e 2023, análise que visa apoiar a decisão do Sr. João, contemplando métricas como:

*   Faturamento total das lojas
*   Categorias de produtos mais e menos vendidas
*   Média das avaliações de clientes por loja
*   Produtos mais e menos vendidos
*   Frete médio de cada loja

## **Material de análise**

Conforme citado, utilizaremos dados de vendas das 4 lojas, referente aos anos de 2020 à 2023. Quanto ao ano de 2023, os dados são parciais e referem-se ao 1º Trimestre, e serão considerados para projeções.

Cada linha do conjunto de dados é referente à uma venda. Cada venda possui os seguintes dados: nome do produto, categoria do produto, preço, frete, data da compra, nome do vendedor, local de compra, avaliação da compra, tipo de pagamento, número de parcelas e geolocalização (latitude e longitude).

Consoante ao frete, que é pago pelo cliente, o valor alto de frete corresponde à vendas mais distantes e produtos mais caros.
